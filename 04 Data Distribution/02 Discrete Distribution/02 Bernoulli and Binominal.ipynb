```python
'''
Example for Binominal Distribution
1. Number of Side Effects from Medications
2. Number of Fraudulent Transactions
3. Number of Spam Emails per Day
4. Number of River Overflows
5. Shopping Returns per Week
```

```python
# get number of outcomes using as above the formula

import math

def get_ncr(n,x):
    return math.factorial(n)/(math.factorial(x) * math.factorial(n-x));  
```

```python
# Finding Binominal with 3 tosses, what are the chances of 2 Heads?

n = 3  # 3 trails
x = 2  # 2 head

get_ncr(n,x)

# So there are 3 outcomes that have "2 Heads"
```

```python
# Example: with 9 tosses, what are the chances of 5 Heads

n = 9
x = 5

get_ncr(n,x)

# So 126 of the outcomes will have 5 heads
```

```python
# Calculate Probability of event = (Number of outcomes we want)  * (Probability of each outcome)

# Example: You sell sandwiches. 70% of people choose chicken, the rest choose something else.
# What is the probability of selling 2 chicken sandwiches to the next 3 customers?

p = 0.7  # % of people chicken choosen
n = 3 # customers 
x = 2 # selling chicken

# find the Number of outcomes we want

ncr = get_ncr(3,2)
print("ncr:",ncr)

# find Probability of each outcome
# px * (1-q)n-x 
Probability_each_Outcome = np.power(p,x) * np.power((1 -p),(n-x))

output = ncr* Probability_each_Outcome
print(output)

#So the probability of event "2 people out of 3 choose chicken" = 0.441
```

```python
## Sam says "70% choose chicken, so 7 of the next 10 customers should choose chicken" ... what are the chances Sam is right?

p = 0.7
n = 10
x = 7

probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))

print(probability)

# So the probability of 7 out of 10 choosing chicken is only about 27%
```

```python
# Example 1: If a coin is tossed 5 times, using binomial distribution find the probability of: 
# (a)Exactly 2 heads (b)At least 4 heads.


#Here, p is 0.5 (2 outcome head and tail)
n = 5
p = 0.5
x = 4

probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))

print(probability)
```

```python
# A coin is tossed 10 times. What is the probability of getting exactly 6 heads?
n = 10
p = 0.5
x = 6

probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))

print(probability)
```

```python
# 80% of people who purchase pet insurance are women.  
# If 9 pet insurance owners are randomly selected, find the probability that exactly 6 are women.

p = 0.8
n = 9
x = 6

probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))

print(probability)
```

```python
# 60% of people who purchase sports cars are men.  
# If 10 sports car owners are randomly selected, find the probability that exactly 7 are men.

p = 0.6
n = 10
x = 7

probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))

print(probability)
```

```python
# Number of Spam Emails per Day
# For example, suppose it is known that 4% of all emails are spam. 
# If an account receives 20 emails in a given day, we can use a Binomial Distribution Calculator to find the 
# probability that a certain number of those emails are spam:

p = 0.04
n = 20
trails = [0,1,2]

for x in trails:
    probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))
    
    print("P(X = {0} Spam Emails = {1})".format(x,np.round(probability,4)))
```

```python
# suppose it is known that a given river overflows during 5% of all storms. 
# If there are 20 storms in a given year, we can use a Binomial Distribution Calculator to find the probability that the 
# river overflows a certain number of

p = 0.05
n = 20
trails = [0,1,2]

for x in trails:
    probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))
    
    print("P(X = {0} Overflow = {1})".format(x,np.round(probability,4)))
```

```python
# Shopping Returns per Week
# Suppose it is known that 10% of all orders get returned at a certain store each week. If there are 50 orders that week
# we can use a Binomial Distribution Calculator to find the probability that the store receives more than a 
# certain number of returns that week:

p = 0.1
n = 50
trails = [5,10,15]

for x in trails:
    probability = get_ncr(n,x) * (np.power(p,x) * np.power((1 -p),(n-x)))
    
    print("P(X > {0} returns = {1})".format(x,np.round(probability,4)))
```

```python
'''
Difference between Bernoulli and Binominal Distribution
    - Bernoulli deals with the outcome of the single trial of the event, whereas Binomial deals with the outcome of the 
      multiple trials of the single event.
      
      Example : Taking Quiz, Guessing 1 question (Yes/No)  is Bernoulli event, and Guessing the entire quiz is Binominal 
                Distribution
                
    - Bernoulli is used when the outcome of an event is required for only one time,whereas the Binomial is used when the 
      outcome of an event is required multiple times
    
```

```python
#Import required libraries:
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

#Define probability of success:
p = 0.7

#Find the statisticsal values:
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')

#Print mean:
print("Mean = ",mean)

#Print variance:
print("Variance =  ",var)

#Print skewness:
print("Skewness = ",skew)

#Print kurtosis:
print("Kurtosis = ",kurt)

#Get only mean value:
mean = bernoulli.mean(p)
print("Mean = ",mean)

#Get only median value:
median = bernoulli.median(p)
print("Median = ",median)

#Get only variance value:
var = bernoulli.var(p)
print("Variance = ",var)

#Get only standard deviation value:
std = bernoulli.std(p)
print("Standard Deviation = ",std)

#Get Probability Mass Function(PMF):
x = [0,1]
p=0.7
print("Probability Mass Function = ",bernoulli.pmf(x,p))

#Plot the graph for Probability Mass Function(PMF):
x = [0,1]
p=0.7
plt.scatter(x,bernoulli.pmf(x,p),label="PMF")
plt.title("Probability Mass Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.legend()

#Get Cumulative Density Function(CDF):
x = [0,1]
p = 0.7
print("Cumulative Density Function = ",bernoulli.cdf(x,p))

#Plot the Cumulative Density Function(CDF):
x = [0,1]
p = 0.7
plt.scatter(x,bernoulli.cdf(x,p),label="CDF")
plt.title("Cumulative Density Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.legend()

#Plot the bar graph for PMF:
x = [0,1]
p = 0.7
plt.bar(x,bernoulli.pmf(x,p),width=0.1,color=["r","b"])
plt.title("Probability Mass Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")


#Plot the bar graph for CDF:
x = [0,1]
p = 0.7
plt.bar(x,bernoulli.cdf(x,p),width=0.1,color=["r","b"])
plt.title("Cumulative Density Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")

#Generate Output for Random Bernoulli Events:
p = 0.7
r = bernoulli.rvs(p, size=100)
print(r)
```
