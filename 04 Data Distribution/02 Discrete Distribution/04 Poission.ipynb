```python
'''
Poission Distribution
    - The frequency distribution of the number of events in sampled units of time or space.
    - Model count data, such as the count of library book checkouts per hour
```

```python
'''
Poission Distribution
    - The Poisson distribution describes the probability of experiencing k events during a fixed time interval.
```

### Poission Distribution



```python
### P(X = k) = P(K events in interval) 
### k = ((events /time) * time period) is usually simplified into a single parameter, λ, lambda
### As we change the rate parameter, λ, we change the probability of seeing different numbers of events in one interval.
```

```python
import numpy as np
import math 
import matplotlib.pyplot as plt

def get_PoissionDist(lamda,k):
    e = 2.7182 # eluer's number
    
    output = (np.power(lamda,k) * np.power(e,-lamda)) / math.factorial(k)
    return np.round(output,3)
```

```python
# Example :  In the World Cup, an average of 2.5 goals are scored each game. Modeling this situation with a Poisson 
# distribution, what is the probability that kk goals are scored in a game?

# In this instance choose lambda = 2.5

_lambda = 2.5

# find the probability for rabdom number X=0,1,2,3,4,5,6

probability =[]
kvalues =  []

for k in range(0,7):
    kvalues.append(k)
    probability.append(get_PoissionDist(2.5,k))

print(probability)

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(kvalues, probability, color ='rgbkymc',
        width = 0.4)
 
plt.xlabel("random numbers")
plt.ylabel("Probability")
plt.show()
```

```python
## Example 1: In a cafe, the customer arrives at a mean rate of 2 per min. Find the probability of arrival of 5 customers 
#             in 1 minute using the Poisson distribution formula.

lamda = 2
x = 5

probability = get_PoissionDist(lamda,x)
print(probability)

# The probability of arrival of 5 customers per minute is 3.6%.
```

```python
## Find the mass probability of function at x = 6, if the value of the mean is 3.4.

lamda = 3.4
x = 6

probability = get_PoissionDist(lamda,x)
print(probability)

# The probability of function is 7.2%.
```

```python
## If 3% of electronic units manufactured by a company are defective. 
#  Find the probability that in a sample of 200 units, less than 2 bulbs are defective.

# The probability of defective units p = 3/100 = 0.03
# Mean λ= np = 200 × 0.03 = 6

n = 200 
lamda = 6

# Condistion is less than 2
# output is Probability(k=0) + Probability(k=1) 

x = 0
probability_0 = get_PoissionDist(lamda,x)
print("probability_X_Equal_0:",probability_0)

x = 1
probability_1 = get_PoissionDist(lamda,x)
print("probability_X_Equal_1:",probability_1)


probability_X_less_than_2 = probability_0 + probability_1

print("probability_X_less_than_2:",probability_X_less_than_2)

# The probability that less than 2 bulbs are defective is 0.017
```

```python
# A store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day? 

from scipy.stats import poisson

#calculate probability
poisson.pmf(k=5, mu=3)

# The probability that the store sells 5 apples in a given day is 0.100819.
```

```python
# A certain store sells 7 footballs per day on average. What is the probability that this store sells 4 or 
# less footballs in a given day?

from scipy.stats import poisson

#calculate probability
poisson.cdf(k=4, mu=7)

# The probability that the store sells four or less footballs in a given day is 0.172992
```

```python
# A certain store sells 15 cans of tuna per day on average. 
# What is the probability that this store sells more than 20 cans of tuna in a given day?

from scipy.stats import poisson

#calculate probability
1 - poisson.cdf(k=20, mu=15)

# The probability that the store sells more than 20 cans of tuna in a given day is 0.082971.
```
