###                                                       Exponential Distribution


```python
'''
Need of Exponential Distribution:
    - To predict the amount of waiting time until the next event (i.e., success, failure, arrival, etc.).
```

```python
'''
Example for Exponential Distribution

1. What is the probability that a customer will arrive within th next 2 minutes
2. What is the probability that a number of calls in next 5 hours
3. Time Between Customers
4. Time Between Earthquakes 
```

```python
'''
Difference between Poission and Exponential Distriution
If the number of events per unit time follows a Poisson distribution, 
then the amount of time between events follows the exponential distribution.
```

```python
import numpy as np

def find_ExponentialDist(lamda,x):
    e = 2.718
    return lamda * np.power(e, -lamda*x)
    
```

```python
# suppose the mean number of minutes between eruptions for a certain geyser is 40 minutes. 
# If a geyser just erupts, what is the probability that well have to wait less than 50 minutes for the next eruption?

# lambda = 1 / mu  

lamda =  np.divide(1,40)
x = 50

find_ExponentialDist(lamda,x)
```
