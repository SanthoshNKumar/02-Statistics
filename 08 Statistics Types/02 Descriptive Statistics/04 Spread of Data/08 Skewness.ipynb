```python
'''
Sweness : It describes the lack of symmetic from the mean. perfect symmetric data will have skewness of 0
Example : Normal Distribution will of skewness of 0

Skewness : 
    - Use skewness to determine the extent(Area) to which the data are not symmetrical.
    
Sweness : It describes the lack of symmetic from the mean. perfect symmetric data will have skewness of 0
          Example : Normal Distribution will of skewness of 0
```

```python
# skewness = 0 : normally distributed.(Symmetic) mean = 0 and Std = 1
# skewness > 0 : more weight in the left tail of the distribution.
# skewness < 0 : more weight in the right tail of the distribution.
```

```python
from scipy.stats import skew 
import numpy as np  
import pylab as p  
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
```

## Skew formula

```python
skew = (3* (np.mean(x1) - np.median(x1))) / np.std(x1)
print(skew)
```

```python
from scipy.stats import skew 
import numpy as np  
import pylab as p  
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
p.plot(x1, y1, '*') 
  
print( '\nSkewness for data : ', skew(x1)) 
```

```python
#Skewness for normally distributed data (Page 86):

#Import required libraries:
from statistics import mean,median,mode
from scipy.stats import skew
import numpy as np

#Get normally distributed data:
X = np.random.normal(0,2,10000)

#Print the data:
print(X,"\n")

#Print Mean:
print("Mean =",mean(X))

#Print Median:
print("Median = ",median(X))

#Print Mode if available:
try:
    print("Mode = ",mode(X))
except:
    print("Mode not found!")
    
#Find the skewness:
print("Skewness =",skew(X))
```
