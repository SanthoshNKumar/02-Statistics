### Confidence Interval for a Mean

```python
'''
# Example 01:

Sample size n = 25
Sample mean weight x = 300
Sample standard deviation s = 18.5

Here is how to find various confidence intervals for the true population mean weight:

90% Confidence Interval: 300 +/-  1.645*(18.5/25) = [293.91, 306.09]
95% Confidence Interval: 300 +/-  1.96*(18.5/25) = [292.75, 307.25]
99% Confidence Interval: 300 +/-  2.58*(18.5/25) = [290.47, 309.53]

## Note : 1.645; 1.96;2.58 are the values from Z table for (90%,95% and 99%)

# There is a 95% chance that the confidence interval of [292.75, 307.25] contains the true population mean weight of turtles
```

### Confidence Interval for the Difference Between Means

```python
# Example 02 : 

'''
Confidence interval = (x1x2) +/- t*((sp2/n1) + (sp2/n2))

x1, x2: sample 1 mean, sample 2 mean
t: the t-critical value based on the confidence level and (n1+n2-2) degrees of freedom
sp2: pooled variance
n1, n2: sample 1 size, sample 2 size


Sample 1:

x1 = 310
s1 = 18.5
n1 = 15
Sample 2:

x2 = 300
s2 = 16.4
n2 = 15

90% Confidence Interval:
(310-300) +/- 1.70*((305.61/15) + (305.61/15)) =  [-0.8589, 20.8589]

95% Confidence Interval:
(310-300) +/- 2.05*((305.61/15) + (305.61/15)) =  [-3.0757, 23.0757]

99% Confidence Interval:
(310-300) +/- 2.76*((305.61/15) + (305.61/15)) =   [-7.6389, 27.6389]

# There is a 95% chance that the confidence interval of [-3.0757, 23.0757] contains the true difference in mean weight 
# between the two turtle populations.
```

### Confidence Interval for a Proportion

```python
# Example 03:

'''
Confidence Interval = p  +/-  z*p(1-p) / n

p: sample proportion
z: the chosen z-value
n: sample size

Sample size n = 100
Proportion in favor of law p = 0.56

Here is how to find various confidence intervals for the population proportion:

90% Confidence Interval: 0.56  +/-  1.645*(.56(1-.56) / 100) = [0.478, 0.642]

95% Confidence Interval: 0.56  +/-  1.96*(.56(1-.56) / 100) = [0.463, 0.657]

99% Confidence Interval: 0.56  +/-  2.58*(.56(1-.56) / 100) = [0.432, 0.688]

# There is a 95% chance that the confidence interval of [0.463, 0.657] contains the true population proportion of 
# residents who are in favor of this certain law.
```

### Confidence Interval for a Standard Deviation

### How to Find a Confidence Interval for a Median

```python
# Example 04: 

'''
j: nq    znq(1-q)

k: nq  +  znq(1-q)

where:

n: The sample size
q: The quantile of interest. For a median, we will use q = 0.5.
z: The z-critical value

Sample data: 8, 11, 12, 13, 15, 17, 19, 20, 21, 21, 22, 23, 25, 26, 28

Step 1: Find the Median
    First, we need to find the median of the sample data. This turns out to be the middle value of 20:

    8, 11, 12, 13, 15, 17, 19, 20, 21, 21, 22, 23, 25, 26, 28

Step 2: Find j and k
    Suppose we would like to find a 95% confidence interval for the population median. To do so, we need to first find j and k:

    j: nq  znq(1-q) = (15)(.5)  1.96(15)(.5)(1-.5) = 3.7
    k: nq + znq(1-q) = (15)(.5) + 1.96(15)(.5)(1-.5) = 11.3
    We will round both j and k up to the nearest integer:

    j: 4
    k: 12
    
Step 3: Find the Confidence Interval

The 95% confidence interval for the median will be between the j = 4th and k = 12th observation in the sample dataset.

The 4th observation is equal to 13 and the 12th observation is equal to 23:
8, 11, 12, 13, 15, 17, 19, 20, 21, 21, 22, 23, 25, 26, 28

# Thus, the 95% confidence interval for the median turns out to be [13, 23].
```

### Confidence Interval for a Correlation Coefficient

```python
# Example 05 : 

'''
Step 1:  Perform Fisher transformation.
    Let zr = ln(1+r / 1-r) / 2

Step 2: Find log upper and lower bounds.
    Let L = zr    (z1-/2 /(n-3))
    Let U = zr  +  (z1-/2 /(n-3))

Step 3: Find confidence interval.
    Confidence interval = [(e2L-1)/(e2L+1),  (e2U-1)/(e2U+1)]
    
Example:
Sample size n = 30
    Correlation coefficient between height and weight r = 0.56
    Here is how to find a 95% confidence interval for the population correlation coefficient:

Step 1:  Perform Fisher transformation.

    Let zr = ln(1+r / 1-r) / 2 = ln(1+.56 / 1-.56) / 2 = 0.6328

Step 2: Find log upper and lower bounds.

    Let L = zr    (z1-/2 /(n-3)) = .6328    (1.96 /(30-3)) = .2556

    Let U = zr  +  (z1-/2 /(n-3)) = .6328  +  (1.96 /(30-3)) = 1.01

Step 3: Find confidence interval.

    Confidence interval = [(e2L-1)/(e2L+1),  (e2U-1)/(e2U+1)] 
    
Confidence interval = [(e2(.2556)-1)/(e2(.2556)+1),  (e2(1.01)-1)/(e2(1.01)+1)] = [.2502, .7658]
```

### Confidence Intervals Using Z Value

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

height_data = np.array([186.0,180.0,195.0,189.0,191.0,177.0,161.0,
                        177.0,192.0,182.0,185.0,192.0,173.0,172.0,
                        191.0,184.0,193.0,182.0,190.0,185.0,181.0,
                        188.0,179.0,188.0,170.0,179.0,180.0,189.0,
                        188.0,185.0,170.0,197.0,187.0,182.0,173.0,
                        179.0,184.0,177.0,190.0,174.0,203.0,206.0,
                        173.0,169.0,178.0,201.0,198.0,166.0,171.0,180.0])

plt.hist(height_data,30)
plt.show()

# Mean of the distribution
height_mean = np.mean(height_data)
print("Height_Mean=",height_mean)

height_sem = stats.sem(height_data)

# Calculate Standard Error
print("Standard Error = ",height_sem)

# We got the Standard Error of 1.38 and we need to calculate upper and lower limit of 

# A 1.96 standard deviation covers 95% of area in the normal distribution.

# z_value for 95% of area in the normal distribution = 1.96

# Upper/Lowerlimit = mean(height)+/-Z_value*SEM 

# For lower limit
lower_limit = height_mean + (1.96 * height_sem)
upper_limit = height_mean - (1.96 * height_sem)

print("Interval limits")
print("Upper Limit:",upper_limit)
print("Lower Limit:",lower_limit)

print("Now we can confidently say that population mean lies between 180.53 and 185.95")

# Here we can confidently say that population mean lies between 180.53cm and 185.94 of height
```

### Confidence Intervals Using the T Distribution

```python
#  Small sample (n <30), we can use the t.interval() function from the scipy.stats library

import numpy as np
import scipy.stats as st

# calculate a confidence interval for the true population mean height (in inches) of a certain species of plant, 
# using a sample of 15 plants:

#define sample data
data = [12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29]

#create 95% confidence interval for population mean weight
print(st.t.interval(alpha=0.95, df=len(data)-1, loc=np.mean(data), scale=st.sem(data)))

# The 95% confidence interval for the true population mean height is (16.758, 24.042).

#create 99% confidence interval for same sample
print(st.t.interval(alpha=0.99, df=len(data)-1, loc=np.mean(data), scale=st.sem(data)))

# Notice that this interval is wider than the previous 95% confidence interval
```

### Confidence Intervals Using the Normal Distribution

```python
# If were working with larger samples (n30), 
# we can assume that the sampling distribution of the sample mean is normally distributed

# Example :  calculate a confidence interval for the true population mean height (in inches) of a certain species of plant,
# Using a sample of 50 plants:

import numpy as np
import scipy.stats as st

#define sample data
np.random.seed(0)
data = np.random.randint(10, 30, 50)

#create 95% confidence interval for population mean weight
st.norm.interval(alpha=0.95, loc=np.mean(data), scale=st.sem(data))

#The 95% confidence interval for the true population mean height is (17.40, 21.08)

# heres how to calculate a 99% C.I. for the exact same data:

#create 99% confidence interval for same sample
st.norm.interval(alpha=0.99, loc=np.mean(data), scale=st.sem(data))

# The 95% confidence interval for the true population mean height is (17.82, 21.66).
```

### Confidence Interval_Boot Strap_Confidence Interval

```python
## How to Calculate Bootstrap Confidence Intervals For Machine Learning Results in Python

import numpy as np
import pandas as pd
from sklearn.utils import resample
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as pyplot

data = pd.read_csv("pima-indians-diabetes.csv",header=None)

values = data.values

#Configure BootStrap
n_iteration = 10 # But Generally it will be 1000
n_size = int(len(data) * 0.50)

#Run Boostrap
stats = list()

for i in range(n_iteration):
    
    #Prepare for Train and Test Sets
    train = resample(values,n_samples = n_size)
    test = np.array([x for x in values if x.tolist() not in train.tolist()])
    
    #fit the Model
    model = DecisionTreeClassifier()
    model.fit(train[:,:-1],train[:,-1])
    
    #Evaluate the Model
    predictions = model.predict(test[:,:-1])
    score = accuracy_score(test[:,-1],predictions)
    print(score)
    stats.append(score)
# plot scores
pyplot.hist(stats)
pyplot.show()

# confidence intervals
alpha = 0.95
p = ((1.0-alpha)/2.0) * 100
lower = max(0.0, np.percentile(stats, p))
p = (alpha+((1.0-alpha)/2.0)) * 100
upper = min(1.0, np.percentile(stats, p))
print('%.1f confidence interval %.1f%% and %.1f%%' % (alpha*100, lower*100, upper*100))
```
