```python
# Variance :Variance measures how far a set of data is spread out.
# A variance of zero indicates that all of the data values are identical
```

### Manual

```python
import numpy as np

x = np.array([1,2,2,3]) # Sample Elements

xbar = np.mean(x) # Mean of the Elemets

n = len(x) # Number of Samples

# Calculate By Using above the Formula
variance = np.divide(np.sum((x-xbar)**2),(n-1))

print("variance:",variance)
print("Standard Deviation:",np.sqrt(variance))
```

### List Comprehension

```python
# Caluclate using List Comprehension
variance = np.sum([(i - np.mean(x)) **2 for i in x]) / (n -1)

print("variance:",variance)
print("Standard Deviation:",np.sqrt(variance))
```

### Numpy

```python
# variance
print("variance:",np.var(x,ddof=1))

# Standard Deviation is the squreroot of variance
print("Standard Deviation:",np.sqrt(variance))

print("Standard Deviation:",np.std(x))
```

### Example 01

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.array([150.5,170.0,160.0,161.0,170.5]),columns=['X'])
data['Deviation(X-X_mean)'] =  data - np.mean(data)
data['Squared Deviation'] = np.square(data['Deviation(X-X_mean)'])


print(data)

n = len(data)

print("Number of Observation:",n)

Sum_of_SquaredDeviation = np.sum(data['Squared Deviation'])
print("Sum of Squared Deviation :",Sum_of_SquaredDeviation)

# Standard Deviation  = SquareRoot(Sum of Squared Deviation / (n -1))

std_Dev = np.sqrt(Sum_of_SquaredDeviation / (n -1))
print("Standard Deviation:",std_Dev)

# std_error =  Standard Deviation / SqreRoot(n)

std_error = std_Dev / np.sqrt(n)
print("Standard Error:",std_error)
```
