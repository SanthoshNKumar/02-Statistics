```python
'''

Kurthosis: 
    - It is a measure of how much heavily the tails of a distribution differs from the tails of a normal distribution.
    - It simply identifies whether the tails of the given distribution contains extreme values.
    
Positive Kurthosis:
    - Positive value of kurthosis means that your dataset is more "peak-heavy"(more peaked) than the normal distribution

Negative Kurthosis:
    - Negative value of kurthosis means your dataset is less 'tail-heavly'(flatter) than the normal distribution
    
A Perfect normal distribution has a kurthosis coefficient of Zero
```

### Formula

```python
from scipy.stats import kurtosistest 
import numpy as np  
import pylab as p  

# null hypothesis  = Test whether a dataset has normal kurtosis.
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
p.plot(x1, y1, '*') 
  
print( '\nKurtosis for normal distribution :\n', kurtosistest(y1)) 

k, p = kurtosistest(y1)

if p < 0.05:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis cannot be rejected")
```

```python
#Mesokurtic (Page 95):
#Fisher=False

#Import required libraries:
from scipy.stats import kurtosis
import numpy as np

#Generating normally distributed data:
X=np.random.normal(0,1,100000)

#Get the kurtosis value:
print("Kurtosis = ",kurtosis(X,fisher=False))
```
