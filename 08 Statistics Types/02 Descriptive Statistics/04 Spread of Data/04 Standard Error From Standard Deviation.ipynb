```python
import pandas as pd
import numpy as np

data = pd.DataFrame(np.array([150.5,170.0,160.0,161.0,170.5]),columns=['X'])
data['Deviation(X-X_mean)'] =  data - np.mean(data)
data['Squared Deviation'] = np.square(data['Deviation(X-X_mean)'])

data
```

```python
n = len(data)

print("Number of Observation:",n)

Sum_of_SquaredDeviation = np.sum(data['Squared Deviation'])
print("Sum of Squared Deviation :",Sum_of_SquaredDeviation)

# Standard Deviation  = SquareRoot(Sum of Squared Deviation / (n -1))

std_Dev = np.sqrt(Sum_of_SquaredDeviation / (n -1))
print("Standard Deviation:",std_Dev)

# std_error =  Standard Deviation / SqreRoot(n)

std_error = std_Dev / np.sqrt(n)
print("Standard Error:",std_error)
```
