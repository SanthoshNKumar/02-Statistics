### Manual Calulation

```python
import numpy as np
import matplotlib.pyplot as plt

# Example :1

data1 = np.array(20 * np.random.randn(1000) + 100)
data2 =np.array(data1 + (10 * np.random.randn(1000) + 50))

n = len(data1)

# Step 1: Find Covariance of two varobes
# Step 2: The Pearson’s correlation coefficient is calculated as the covariance of 
# the two variables divided by the product of the standard deviation of each data sample.

# Calculate Covariance

cov =np.sum((data1 - np.mean(data1)) * (data2 - np.mean(data2))) / (n -1)

print(cov)
```

```python
# Calculate Correlation Coefficent
r = cov / (np.std(data1) * np.std(data2))
print(r)
```

### Using Numpy

```python
# using Numpy
# Step 1: Calculate Covariance
covv = np.cov(data1,data2)
print(covv)

# Step 2: Calculate Correlation Coefficents
Stddata1 = np.std(data1)
Stddata2 = np.std(data2)


r = covv / (Stddata1 * Stddata2)

r
```

```python
# Direct Caluclate Correlation using Numpy
np.corrcoef(data1,data2)
```
