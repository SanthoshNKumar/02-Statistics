```python
import numpy as np
import pandas as pd

graduation_data = {'years_after_starting_college': [2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7], 
                   'count_students_graduated': [5, 50, 150, 3380, 760, 340, 115, 80, 40, 10],
                   'count_students_graduated_running_total' :[5,55,205,3585,4345,4685,4800,4880,4920,4930]}
df = pd.DataFrame(data=graduation_data)
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(x='years_after_starting_college',y='count_students_graduated_running_total', kind='line',figsize=(10,8),legend=False,
        style='yo-',label="count students graduated running total")
plt.title("Running Total of Students Who Graduated in $X$ Years\nFrom Same Start Class", y=1.01, fontsize=20)
plt.ylabel("running total of students graduated", labelpad=15)
plt.xlabel("years after starting college", labelpad=15)
plt.legend();
```

```python
df['percent_students_graduated_of_total'] = (df['count_students_graduated_running_total'] \
                                             / sum(df['count_students_graduated'])) * 100
```

```python
# A cumulative relative frequency graph of a quantitative variable is a curve graphically showing the cumulative
# relative frequency distribution.
```

```python
df.plot(x='years_after_starting_college',y='percent_students_graduated_of_total', kind='line',figsize=(10,8),legend=False,
        style='yo-',label="count students graduated running total")
plt.title("Running Total of Students Who Graduated in $X$ Years\nFrom Same Start Class", y=1.01, fontsize=20)
plt.ylabel("running total of students graduated", labelpad=15)
plt.xlabel("years after starting college", labelpad=15)
plt.legend();
```
