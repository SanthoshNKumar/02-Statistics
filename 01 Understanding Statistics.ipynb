```python
#####  Probability

'''
Probability: Probability is the measure of the likelihood that an event will occur in a Random Experiment

Probability of A and B : The probability of A and B means that we want to know the probability of two events happening at the 
same time.

P(A) = Probability of event A

P(B) = Probability of event B

P(B|A) = Conditional probability of event B given that event A

Joint Probability: Two events happening together. Example : P(A) ? P(B)

P(A) ? P(B) = P(A) or P(B)  = (Union of two Sets)

P(A) n P(B) = P(A) and P(B) = (Intersection of Two Sets)

p(A and B) (independent events) = p(A) * p(B).

p(A and B) (dependent events) =   p(A) * p(B|A)

p(A) * p(B|A)  = the probability of event A times the probability of event B given event A

Bayes’ Theorem : Bayes' Theorem is a way of finding a probability when we know certain other probabilities.
         Bayes’s theorem is a relationship between the conditional probabilities of two events
         P(A|B) =  P(A) P(B|A)P(B)
              - P(A|B) : Posterior Probability of 'A' given the evidence
              
Conditional Probability:P(A|B) 
    - It is a measure of the probability of one event occurring with some relationship to one or more other events.

Independent Events:
    - Two events are independent if the occurrence of one does not affect the probability of occurrence of the other.

Bayes’ Theorem
    - Bayes’ Theorem describes the probability of an event based on prior knowledge of conditions that might be related to the 
      event.

Mutually Exclusive Events:
    - Two events are mutually exclusive if they cannot both occur at the same time. 
```

```python
##### Null Hypothesis
'''
ANOVA  : Analysis of Variance
    - Null hypotheses: Groups means are equal (no variation in means of groups)
    - Alternative hypotheses: At least, one group mean is different from other groups

Correlation Test
    - Null Hypothesis : there is correlation/association between two variables in a population
    - Alternative Hypothesis :  there is no correlation/association between two variables in a population
    
Pearson’s Correlation Coefficient
     - Null Hypothesis :the two samples are independent. (No Correlation between samples)
     - Alternate Hypothesis :there is a dependency between the samples. (there is a Correlation between samples)
     
Shapiro-Wilk Test : Test data sample is Gaussian distribution
    - Null Hypothesis : the sample has a Gaussian(Normal) distribution.
    - Alternative Hypothesis :t he sample does not have a Gaussian distribution.

Anderson-Darling Test : Test data sample is Gaussian distribution
    - Null Hypothesis : the sample has a Gaussian distribution.
    - Alternate Hypothesis : the sample does not have a Gaussian distribution.
    
One sample t-test :
    - Null Hypothesis : Sample Mean and population Mean are same
    - Alternate Hypothesis : Sample Mean and Popluation Mean are differs
    
Two sample t-test (Student's t-tests): compare mean of two groups
    - Null Hypothesis: Means difference between two sample is 0 (Mean of two different groups are same)
    - Alternate Hypothesis: Mean difference between two sample is not 0 (Mean of two different groups are different)
    
Paired t test: compare mean from same group at different times
    - Null Hypothesis: Means difference between two sample is 0 (Mean of two different groups are same)
    - Alternate Hypothesis: Mean difference between two sample is not 0 (Mean of two different groups are different)
    
One sample Z-test: (Use Sample size if greater than 30)
    - Null Hypothesis: Sample Mean and population Mean are same
    - Alternate Hypothesis: Sample Mean and Popluation Mean are differs

Two sample Z-test:
    - Null Hypothesis: mean of two group is 0 (mean of different groups are same)
    - Alternate Hypothesis: mean of two group is not 0(mean of different groups are not same)
    
F Test : If F value is greater than F-Statistics or F-critical value we reject the null hypothesis
    - Null Hypothesis : All the beta values(Coefficents) are equal to Zero
    - Alternative Hypothesis : Alteast one Beta Value differes from the zero
    
Skew Test : test the Normality of data
    - Null Hypothesis: The data is normally distributed.
    - Alternative Hypothesis:The data is not normally distributed
    
Two tailed Test : 
    - Null Hypothesis: Difference between the true mean (μ) and the comparison value is equal to zero. 
    - Alternative Hypothesis: Difference between the true mean (μ) and the comparison value is not equal to zero.
    
One tailed Test (Upper-tailed):
    - Null Hypothesis: Difference between the true mean (μ) and the comparison value is equal to zero.
    - Alternative Hypothesis:Assumes that the true mean (μ) of the sample is greater than the comparison value(m0)
    
One tailed Test (Lower-tailed):
    - Null Hypothesis: Difference between the true mean (μ) and the comparison value is equal to zero.
    - Alternative Hypothesis: Assumes that the true mean (μ) of the sample is less than the comparison value (m0).

Chi-Squared Test :(Non Parametric Test) Goodness of Fit
    - Null Hypothesis: Given variable(Categorical) is likely to come from a given or population distribution
    - Alternative Hypothesis : Given variable(Categorical) is not likely to come from a given or population distribution
    
Chi-Squared Test :(Non Parametric Test) Test independent
    - Null Hypothesis : No relationship exists on the categorical variables in the population; they are independent. 
    - Alternative Hypothesis : There are relationships between the categorical variable

Wilcoxon Signed-Rank Test: (Non Parametric Test) :- Equivalent to (One Sample Z-Test and T-Test)
    - Null Hypothesis: the distributions of both samples are equal.
    - Alternate Hypothesis: the distributions of both samples are not equal.
    
Mann-Whitney U Test : (Non Parametric Test) :- Equivalent to (2 Sample T-Test) or (Un paired T-Test)
    - Null Hypothesis :The distributions of both samples are equal.
    - Alternative Hypothesis : the distributions of both samples are not equal.

Kruskal-Wallis H Test:(Non Parametric Test) :- Equivalent to (One-Way ANOVA)
    - Null Hypothesis: the distributions of both samples are equal.
    - Alternate Hypothesis: the distributions of both samples are not equal.
    
Friedman Test: (Non Parametric Test) :- Equivalent to (Two-Way ANOVA)
    - Null Hypothesis: the distributions of both samples are equal.
    - Alternate Hypothesis: the distributions of both samples are not equal.

Regression : Determine whether there is a significant linear relationship between an independent variable X and a dependent 
             variable Y. Y = B0 + B1X, B0 = Coefficent B1 = Slope
    - Null Hypothesis: The Slope is equal to zero               Ho: Β1 = 0
    - Alternative Hypothesis: The slope is not Equal to zero    Ha: Β1 ≠ 0
```
