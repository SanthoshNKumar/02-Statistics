### Calculate Z -score Manully

### One Sample Z-Test

```python
import pandas as pd
from scipy import stats
df = pd.read_csv("blood_pressure.csv")

print(df.head())

# Example again we are using z-test for blood pressure with some mean like 156 (python code is below for same) 
# one-sample Z test.

import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

## Positive Case

ztest ,pval = stests.ztest(df['bp_before'], x2=None, value=156)
print(float(pval))

if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
    
print(df['bp_before'].mean())

```

```python
## Negative Case:

ztest ,pval = stests.ztest(df['bp_before'], x2=None, value=170)
print(float(pval))

if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
    
print(df['bp_before'].mean())
```

### Two Sample Z-Test

```python
# similar to t-test here we are checking two independent data groups and deciding whether 
# sample mean of two group is equal or not.


# Null Hypothesis :mean of two group is 0 (mean of different groups are same)
# Alternate Hypothesis : mean of two group is not 0(mean of different groups are not same)


import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

df = pd.read_csv("blood_pressure.csv")

# Positive case:
ztest ,p = stests.ztest(df['bp_before'], x2=df['bp_after'], value=0,alternative='two-sided')

print(float(p))
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
    
print("Mean of datset before and after is {0} and {1} different rejecting null hypothesis ".format( \
                                df['bp_before'].mean(),df['bp_after'].mean()))
```

```python
# Negetaive case:

ztest ,p = stests.ztest(df['bp_before'], x2=df['bp_before'], value=0,alternative='two-sided')

print(float(p))
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
    
print("Mean of datset before and after is {0} and {1} are same so accepting null hypothesis ".format( \
                                df['bp_before'].mean(),df['bp_before'].mean()))
```
