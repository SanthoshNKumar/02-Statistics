```python
'''
T test : It is a type of inferential statistic which is used to determine if there is a significant difference between the 
         means of two groups which may be related in certain features.
```

```python
'''
T-test Types
    - One sample t-test:
    - One sample t-test:
    - Paired t test:
    
One sample t-test : Which tests the mean of a single group against a known mean.
     Example :- you have 10 ages and you are checking whether avg age is 30 or not. 
     
Two sampled T-test(Independent samples t-test) :  Which compares mean for two groups.
    Example : is there any association between week1 and week2 
    
Paired sampled t-test(dependent sample t-test) : Which compares means from the same group at different times.
    Example: Collect the blood pressure for an individual before and after some treatment, condition, or time point.
```

## One sample T Test

### One Sample T Test Manually

```python
# Example

import numpy as np
blood_pressure = np.array([128,118,144,133,132,111,149,139,136,126,127,115,142,140,131,132,122,119,129,128])

# null hypothses = mean of blood presure  is equal to 120 (population mean)
# Alternate hypothesis = mean od sample is not equal to 120 (population mean)

# Condition : if the calculated t_statistic value is greater than  t_critical then we reject the null hypothesis

x_bar =  np.mean(blood_pressure) # Mean of the input sample
n = len(blood_pressure) # number of observation
s = np.std(blood_pressure) # Standard Deviation
df = (n-1) # Degree of freedom

t_statistic = (x_bar - 120) / (s / np.sqrt(n))

t_statistic
```

```python
# Get the t_critical value from the t table ( here we have look for df of 19 and alpha =0.05)
from scipy.stats import t

alpha = 0.05
p = 1- alpha   # probability

# retrieve value <= probability

t_critical = t.ppf(p, 19)

print(t_critical)

if t_statistic > t_critical:
    print("Here t_statistic > t_critical so Reject Null hypothesis")
else:
    print("Accept null hypothesis")
```

### Example 02

```python
# Example for One Sample T Test

from scipy.stats import ttest_1samp
import numpy as np

blood_pressure = np.array([128,118,144,133,132,111,149,139,136,126,127,115,142,140,131,132,122,119,129,128])

# null hypothses = mean of blood presure is equal to 120 (population mean)
# Alternate hypothesis = mean od sample is not equal to 120 (population mean)

ages_mean = np.mean(blood_pressure)
print(blood_pressure)
tset, pval = ttest_1samp(blood_pressure, 120)

print("pval:",pval)

if pval < 0.05:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis")
else:
  print("we are accepting null hypothesis")
```

### Example 03

```python
# One sample t-test :

# The One Sample t Test determines whether the sample mean is statistically different from a known or 
# hypothesised population mean. 
# The One Sample t Test is a parametric test.


# Example :- you have 10 ages and you are checking whether avg age is 30 or not

import numpy as np
from scipy.stats import ttest_1samp

ages = np.array([32,34,29,29,22,39,38,37,38,36,30,26,22,22])

print(ages)

ages_mean = np.mean(ages)
print("Mean : =",ages_mean)

tset, pval = ttest_1samp(ages, 30)

print("P-valus: =",pval)

if pval < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

### Two Sample T Test Manually

```python
'''
The Independent Sample t Test or Two-sample t-test compares the means of two independent groups in order to determine whether 
there is statistical evidence that the associated population means are significantly different.
```

```python
# Example for Two Sample t Test
# Example : is there any association between week1 and week2

from scipy.stats import ttest_ind
import numpy as np
import pandas as pd

week1 = pd.read_csv("week1.csv")
week2 = pd.read_csv("week2.csv")

week1_mean = np.mean(week1)
week2_mean = np.mean(week2)

print("week1 mean value:",week1_mean)
print("week2 mean value:",week2_mean)

week1_std = np.std(week1)
week2_std = np.std(week2)

print("week1 std value:",week1_std)
print("week2 std value:",week2_std)

print("\n")
ttest,pval = ttest_ind(week1,week2)

print("pval:",pval)

if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
```

### Paired T test

```python
# Paired t test : The paired sample t-test is also called dependent sample t-test
# It is uni variate test that tests for a significant difference between 2 related variables.

# Example : Collect the blood pressure for an individual before and after some treatment


# NUll Hypothesis : means difference between two sample is 0 (Mean of two different groups are same)
# Alternative Hypothses : mean difference between two sample is not 0 (Mean of two different groups are same)
```

```python
# Paired Studentâ€™s t-test
#Tests whether the means of two paired samples are significantly different.

# Assumptions
# Observations in each sample are independent and identically distributed (iid).
# Observations in each sample are normally distributed.
# Observations in each sample have the same variance.
# Observations across each sample are paired.

# Null Hypothesis: the means of the samples are equal.
# Alternative Hypothesis: the means of the samples are unequal.
```

### Example 01

```python
# Paired sampled t-test :

import pandas as pd
from scipy import stats

df = pd.read_csv("blood_pressure.csv")
df[['bp_before','bp_after']].describe()

ttest,pval = stats.ttest_rel(df['bp_before'], df['bp_after'])

print("pval:",pval)

if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
