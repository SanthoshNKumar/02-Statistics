### 01 QQ and Hisogram Plot

### 02 Anderson-Darling Test

```python
# Anderson-Darling Test
# Tests whether a data sample has a Gaussian distribution.


# Assumptions
# 1. Observations in each sample are independent and identically distributed (iid).

# Null Hypothesis : the sample has a Gaussian distribution.
# Alternate Hypothesis : the sample does not have a Gaussian distribution.

# Example of the Anderson-Darling Normality Test
from scipy.stats import anderson
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]

result = anderson(data)

print('stat=%.3f' % (result.statistic))

for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Probably Gaussian at the %.1f%% level' % (sl))
    else:
        print('Probably not Gaussian at the %.1f%% level' % (sl))
```

### 03 SkewTest

```python
from scipy.stats import skewtest 
import numpy as np  
import pylab as p 

# null Hypothesis = sample comes from a normal distribution
  
x1 = np.linspace( -5, 12, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
p.plot(x1, y1, '.') 

k, p = skewtest(y1)
  
print( '\nSkewness for data :\n', skewtest(y1)) 

if p < 0.05:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis cannot be rejected")
```

### 03 using Kurthosis

```python
from scipy.stats import kurtosistest 
import numpy as np  
import pylab as p  

# null hypothesis  = Test whether a dataset has normal kurtosis.
  
x1 = np.linspace( -5, 5, 1000 ) 
y1 = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x1)**2  ) 
  
p.plot(x1, y1, '*') 
  
print( '\nKurtosis for normal distribution :\n', kurtosistest(y1)) 

k, p = kurtosistest(y1)

if p < 0.05:  # null hypothesis: x comes from a normal distribution
    print("The null hypothesis can be rejected")
else:
    print("The null hypothesis cannot be rejected")
```

### 04 D'Agostino's K^2

```python
#  D'Agostino's K^2

# Tests whether a data sample has a Gaussian distribution

# Assumptions: Observations in each sample are independent and identically distributed (iid).

# Null Hypothesis : The sample has a Gaussian distribution.
# Alternate Hypothesis : The sample does not have a Gaussian distribution.

# Example of the D'Agostino's K^2 Normality Test
from scipy.stats import normaltest
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]

plt.hist(data)
plt.show()

stat, p = normaltest(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Accept null Hypothesis")
    print('Probably Gaussian')
else:
    print("Reject null Hypothesis")
    print('Probably not Gaussian')
```

### 06 Shapiro-Wilk Test

```python
# Shapiro-Wilk Test :

# statistical tests that you can use to check if your data has a Gaussian distribution.

# Null Hypothesis : the sample has a Gaussian(Normal) distribution.
# Alternative Hypothesis : he sample does not have a Gaussian distribution.

# Example of the Shapiro-Wilk Normality Test
from scipy.stats import shapiro

# data = np.random.rand(100) # mean =0 and std deviation = 1 and size = 100
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]

plt.hist(data)
plt.show()

stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Accept the Null Hypthesis")
    print('Probably Gaussian')
else:
    print("Reject Null Hypthesis")
    print('Probably not Gaussian')
```
