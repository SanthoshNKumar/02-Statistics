```python
import statsmodels.api as sm
import numpy as np

nobs = 100

X = np.random.random((nobs, 2))
X = sm.add_constant(X)

beta = [1, .1, .5]
e = np.random.random(nobs)
y = np.dot(X, beta) + e

results = sm.OLS(y, X).fit()

print(results.summary())
```

```python
dict = {"X":[3,5,7,9,11,13],"Y":[12,8,10,9,5,2]}

import numpy as np
import pandas as pd

df = pd.DataFrame(dict)

df
```

```python
interept = 14.52380952
coefficient1 = -0.857142857

df['Y_Predicted'] = interept + df['X'] * coefficient1
```

```python
# Mean of X
print(np.mean(df['X']))

# mean of 'Y'
print(np.mean(df['Y']))
```

```python
df['Deviation_X'] = df['X'] - np.mean(df['X'])

df['Squared_Deviation_X'] = df['Deviation_X'] * df['Deviation_X']

df['Deviation_Y'] = df['Y'] - np.mean(df['Y'])

df['Residuals'] = df['Y'] - df['Y_Predicted']

df['Squared_Residuals'] = df['Residuals'] * df['Residuals']

df['Deviation_Predicted'] = np.mean(df['Y']) - df['Y_Predicted']

df['Squared_Deviation_Predicted'] = df['Deviation_Predicted'] * df['Deviation_Predicted']
```

```python
# SSR : sum of squares regression
```

```python
df
```

```python
n = len(df['X'])

print("Total number of observation:",n)

print("Sum of Squared Residuals:" ,np.sum(df['Squared_Residuals']))

print("Sum of Squared Regression:" ,np.sum(df['Squared_Deviation_Predicted']))
```

```python
std_deviation = np.sqrt((1/n)*np.sum(df['Squared_Deviation_X']))
print("Standard Diviation:",std_deviation)

std_error = std_deviation / np.sqrt(n)
print("Standard Error:",std_error)
```
